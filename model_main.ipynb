```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
genre_cv=CountVectorizer(max_features=10000)
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
genre_vector=genre_cv.fit_transform(new_data['genre'].values.astype('U')).toarray()
```

```python
print(vector.shape)
print(genre_vector.shape)
```

```python
print(vector)
print('______________________________________')
print(genre_vector)


```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)#36 detik
genre_similarity =cosine_similarity(genre_vector)#36 detik
```

```python
def recommand(movies):
  index = new_data[new_data['title']==movies].index[0]
  distance = sorted(list(enumerate(similarity[index])),reverse=True, key=lambda vector:vector[1])
  a = 1
  for i in distance[:10]:
    print(a, new_data.iloc[i[0]].title)
    print(f'ID: {i[0]}, Akurasi: {i[1]}')
    a +=1


def genre_recommand(movies):
  index = new_data[new_data['title']==movies].index[0]
  genre_distance = sorted(list(enumerate(genre_similarity[index])),reverse=True, key=lambda vector:vector[1])
  a = 1
  for i in genre_distance[:10]:
    print(a, new_data.iloc[i[0]].title)
    print(f'ID: {i[0]}, Akurasi: {i[1]}')
    print(f'Genre: {new_data.iloc[i[0]].genre}')
    a +=1
```

```python
#jika menggabungkan title, genre, dan overview untuk dilakukan similarity
recommand('Naruto Shippuden the Movie')
```

```python
#jika tidak menggabungkan title, genre, dan overview untuk dilakukan similarity
recommand('The Last: Naruto the Movie')
```
